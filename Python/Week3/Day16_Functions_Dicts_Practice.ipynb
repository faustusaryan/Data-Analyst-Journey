```python
# Department Finder
# Ek function get_department(employee_name) likho jo naam lekar department return kare.
# Rules:
#   "Raj" ya "Amit"   →  return "IT"
#   "Priya" ya "Vivek" →  return "HR"
#   Koi bhi aur naam  →  return "Unknown"


def get_dept(emp_name):
    if emp_name == 'Raj' or emp_name == 'Amit':
        return 'IT'
    elif emp_name == 'Priya' or emp_name == 'Vivek':
        return 'HR'
    else:
        return 'Unknown'
```

```python
# Salary Filter Counter
# Ek function count_above_threshold(salary_list, threshold) likho jo count kare ki list mein kitne salaries threshold se zyada hain.
# Input:  [45000, 32000, 58000, 28000, 41000]
#         threshold = 40000

def count_above_threshold(sal_list, threshold):
    i = 0
    for a in sal_list:
        if a > threshold:
            i += 1
    return i 

count_above_threshold([45000, 32000, 58000, 28000, 41000], 40000)
```

```python
# Salary Stats Calculator
# Ek function get_salary_stats(salary_list) likho jo teen values return kare — minimum, maximum, aur average. Average ko round() se 2 decimal tak rakho.
# Input:  [45000, 32000, 58000, 28000, 41000]

def get_salary_stats(sal_list):
    minimum = min(sal_list)
    maximum = max(sal_list)
    average = round(sum(sal_list) / len(sal_list), 2)
    return minimum, maximum, average

get_salary_stats([45000, 32000, 58000, 28000, 41000])
```

```python
# •	D1: Simple dict banao — {name: 'Raj', salary: 45000, dept: 'IT'} — access karo

emp_info = {'name': 'Raj', 'salary': 45000, 'dept': 'IT'}
print(emp_info)
```

```python
# D2 — Add / Update / Delete Keys, D1 wali dict lo. Teen kaam karo ek-ek karke:

# Naya key city = 'Delhi' add karo
# salary ko 50000 se update karo
# dept key delete karo
# Har step ke baad dict print karo taaki change dikh sake.

emp_info.update({'city': 'Delhi'})
print('after adding city :', emp_info)
emp_info['salary'] = 50000 
print('after updating salary :', emp_info)
emp_info.pop('dept')
print('after deleting dept :', emp_info)
```

```python
# D3 — .keys() / .values() / .items()
# Ek employee dict lo (D1 wali). Teeno methods use karo aur output alag-alag print karo:
# Sirf saari keys
# Sirf saari values
# Key-value pairs saath mein

print(emp_info.keys())
print(emp_info.values())
print(emp_info.items())
```

```python
# D4 — Loop Over Dict
# Neeche diye employees ki list hai. Loop lagao aur format mein print karo:
# Raj ki salary hai: 45000
# Priya ki salary hai: 38000
# Amit ki salary hai: 52000

Data = {'Raj': 45000, 'Priya': 38000, 'Amit': 52000}
for key, val in Data.items():
    print(key, 'ki salary hai:', val)
```

```python
# D5 — Nested Dict Banao
# Ek nested dictionary banao jisme 3 employees ho — emp1, emp2, emp3. Har employee ka name, salary, aur dept ho.
# Dict banake poori print karo.

d = {
    'emp1': {
        'name': 'Aryan',
        'salary': 85000,
        'dept': 'AI'
    },       
    'emp2': {
        'name': 'Gauri',
        'salary': 75000,
        'dept': 'Development'
    },
    'emp3': {
        'name': 'Faustus',
        'salary': 95000,
        'dept': 'Data Science'
    }
}
print(d)
```

```python
# D6 — Nested Dict Access
# D5 wali nested dict se sirf emp2 ka naam aur salary access karke print karo. Direct indexing use karo, loop mat lagao.

print('Name =', d['emp2']['name'])
print('Salary =', d['emp2']['salary'])
```

```python
# D7 — Function: Naam Return Karo
# Ek function get_name(employee) banao jo ek employee dict accept kare (jisme name key ho) aur sirf naam return kare.

def get_name(employee):
    return employee['name']
    
emp = {'name': 'Raj', 'salary': 45000} 
get_name(emp)
```

```python
# D8 — Function: Salary Increment
# Ek function give_increment(employee) banao jo dict le aur uski salary mein 10% badha ke wahi dict return kare. 
# Original dict modify honi chahiye.

def give_increment(employee):
    employee['salary'] = round(employee['salary'] * 1.10)
    return employee

emp = {'name': 'Raj', 'salary': 45000}
give_increment(emp)
```

```python
# D9 — Dict Comprehension
# Neeche do lists di hain. Dict comprehension use karke ek dict banao jisme name: salary ho.

names = ['Raj', 'Priya', 'Amit']
salaries = [45000, 38000, 52000]
a = {name: salary for name, salary in zip(names, salaries)}
print(a)
```

```python
# **D10 — Key Exist Karta Hai?**
# Ek employee dict lo. `'in'` operator use karke check karo:
# 1. `'salary'` key exist karti hai?
# 2. `'bonus'` key exist karti hai?
# Dono cases mein ek readable message print karo — exist karta hai ya nahi.

employee = {'name': 'Raj', 'salary': 45000, 'dept': 'AI'}

if 'salary' in employee.keys():
    print('salary key exists in dict')
else:
    print('salary key does not exists in dict')

if 'bonus' in employee.keys():
    print('bonus key exists in dict')
else:
    print('bonus key does not exists in dict')

```

```python
# **D11 — List of Dicts Banao**
# 5 employees ki list of dicts banao. Har dict mein `name`, `salary`, aur `dept` hona chahiye. 
# Kuch IT mein hon, kuch HR mein — D13–D16 ke liye kaam aayega.

new_emp = [{
        'name': 'Aryan',
        'salary': 89000,
        'dept': 'AI'}, 
          {
            'name': 'Gauri',
            'salary': 95000,
            'dept': 'IT'
          },
          {
            'name': 'Arshi',
            'salary': 40000,
            'dept': 'HR'
          },
          {
            'name': 'Faustus',
            'salary': 92000,
            'dept': 'IT'
          },
          {
            'name': 'Teena',
            'salary': 30000,
            'dept': 'HR'
          }]
print(new_emp)
```

```python
# **D12 — List of Dicts Loop**
# D11 wali list lo aur loop lagake format mein print karo:
# Name: Raj | Salary: 45000 | Dept: IT
# Name: Priya | Salary: 38000 | Dept: HR

for a in new_emp:
    print('Name:', a['name'], '|', 'Salary:', a['salary'], '|', 'Dept:', a['dept'])
```

```python
# **D13 — Filter: High Salary Employees**
# D11 wali list se sirf un employees ko filter karo jinki `salary > 40000` ho. 
# Result ek nayi list mein store karo aur print karo. List comprehension use karo.

high_salary = [more for more in new_emp if more['salary'] > 40000]
print(high_salary)
```

```python
# **D14 — Sort By Salary**
# D11 wali list ko `salary` ke basis pe ascending order mein sort karo. `sorted()` with `key=` use karo.
# Sorted list print karo.

sort_list = sorted(new_emp, key=lambda x: x['salary'])
print(sort_list)
```

```python
# **D15 — Max Salary Wala Employee**
# D11 wali list mein se sabse zyada salary wale employee ka **poora record** (dict) dhundho aur print karo.
# `max()` with `key=` use karo.

maxx = max(new_emp, key = lambda x: x['salary'])
print(maxx)
```

```python
# D16 — Department Wise Grouping
# D11 wali list lo aur ek nayi dict banao jisme department naam key ho aur us department ke 
# employees ke naam ki list value ho:

empl = {}

for el in new_emp:
    dept = el['dept']
    if dept not in empl:
        empl[dept] = []
    empl[dept].append(el['name'])

print(empl)
```
