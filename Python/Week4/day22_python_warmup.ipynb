```python
# Exercise 1: Filter Expensive Products

# Problem:
# Given a list of product names and a dictionary of their prices,
# return a list of product names whose price is greater than 1000.

products = ['Laptop', 'Mouse', 'Keyboard', 'Headphones', 'Monitor']
prices = {
    'Laptop': 55000,
    'Mouse': 500,
    'Keyboard': 800,
    'Headphones': 3500,
    'Monitor': 18000
}

new_list =[]
for p,q in prices.items():
    if q > 1000:
        new_list.append(p)
print(new_list)

# Expected Output:
# ['Laptop', 'Headphones', 'Monitor']
```

```python
# Exercise 2: Filter IT Employees

# Problem:
# Given a list of employee dictionaries,
# return only those employees who belong to the 'IT' department.

employees = [
    {'name': 'Rajesh', 'dept': 'IT', 'salary': 75000},
    {'name': 'Sunita', 'dept': 'HR', 'salary': 55000},
    {'name': 'Aakash', 'dept': 'IT', 'salary': 60000},
]

emp = []
for e in employees:
    dept = e['dept']
    if dept == 'IT':
        emp.append(e)
print(emp)

# Expected Output:
# [
#   {'name': 'Rajesh', 'dept': 'IT', 'salary': 75000},
#   {'name': 'Aakash', 'dept': 'IT', 'salary': 60000}
# ]
```

```python
# Exercise 3: Average Salary per Department

# Problem:
# Given a dictionary where keys are departments and values are lists of salaries,
# calculate the average salary for each department.

dept_salaries = {
    'IT': [75000, 60000, 45000],
    'HR': [55000, 42000, 48000]
}

avg_salary = {}
for d, s in dept_salaries.items():
    avg_salary[d] = round(sum(s)/len(s), 2)
print(avg_salary)

# Expected Output:
# {'IT': 60000.0, 'HR': 48333.33}
```

```python
# Exercise 4: Total Amount of Delivered Orders

# Problem:
# Given a list of orders where each order contains an amount and a status,
# calculate the total amount for only those orders whose status is 'Delivered'.

orders = [
    {'id': 1, 'amount': 15000, 'status': 'Delivered'},
    {'id': 2, 'amount': 20000, 'status': 'Pending'},
    {'id': 3, 'amount': 12000, 'status': 'Delivered'},
    {'id': 4, 'amount': 18000, 'status': 'Delivered'},
]

total_amount = 0
for a in orders:
    if a['status'] == 'Delivered':
        total_amount += a['amount']
print(total_amount)

# Expected Output:
# 45000
```

```python
# Exercise 5: Group Customers by City

# Problem:
# Given a list of customers with their city,
# group customer names by their respective cities.

customers = [
    {'name': 'Amit', 'city': 'Delhi'},
    {'name': 'Sneha', 'city': 'Delhi'},
    {'name': 'Ananya', 'city': 'Delhi'},
    {'name': 'Priya', 'city': 'Mumbai'},
    {'name': 'Arjun', 'city': 'Mumbai'},
    {'name': 'Ravi', 'city': 'Bangalore'}
]

new_dict = {}
for c in customers:
    city = c['city']
    name = c['name']
    if city not in new_dict:
        new_dict[city] = [name]
    else:
        new_dict[city].append(name)
print(new_dict)

# Expected Output:
# {
#   'Delhi': ['Amit', 'Sneha', 'Ananya'],
#   'Mumbai': ['Priya', 'Arjun'],
#   'Bangalore': ['Ravi']
# }

# hint lena pda else condition m thodi mistake kr raha tha m
```

```python
# Exercise 6: Find Most Expensive Product

# Problem:
# From the given products and prices,
# return the name of the product with the highest price.

products = ['Laptop', 'Mouse', 'Keyboard', 'Headphones', 'Monitor']
prices = {
    'Laptop': 55000,
    'Mouse': 500,
    'Keyboard': 800,
    'Headphones': 3500,
    'Monitor': 18000
}

max_price = max(products, key=lambda p: prices[p])
print(max_price)
    
# Expected Output:
# 'Laptop'

most_expensive = None
highest_price = 0

for product in products:
    if prices[product] > highest_price:
        highest_price = prices[product]
        most_expensive = product
print(most_expensive)

#failed help leni pdi kafi 
```

```python
# Exercise 7: Combine Two Lists into Dictionary

# Problem:
# Given two lists (names and ages),
# combine them into a dictionary with name as key and age as value.

names = ['Amit', 'Priya', 'Rahul', 'Sneha']
ages = [28, 34, 25, 30]

info = {}
for a in range(len(names)):
    name = names[a]
    age = ages[a]
    if name not in info:
        info[name] = age
print(info)
        
# Expected Output:
# {'Amit': 28, 'Priya': 34, 'Rahul': 25, 'Sneha': 30}
```

```python
# Exercise 8: Extract Name and Salary

# Problem:
# Given a list of employee dictionaries,
# create a new list containing only 'name' and 'salary' fields.

employees = [
    {'name': 'Rajesh', 'dept': 'IT', 'salary': 75000},
    {'name': 'Sunita', 'dept': 'HR', 'salary': 55000},
    {'name': 'Aakash', 'dept': 'IT', 'salary': 60000},
]

new_dict = []
for e in employees:
    name = e['name']
    salary = e['salary']
    new_dict.append({'name': name, 'salary': salary})
print(new_dict)

# Expected Output:
# [
#   {'name': 'Rajesh', 'salary': 75000},
#   {'name': 'Sunita', 'salary': 55000},
#   {'name': 'Aakash', 'salary': 60000}
# ]
```

```python
# Exercise 9: Remove Duplicates (Maintain Order)

# Problem:
# Given a list of cities with duplicates,
# return a new list with duplicates removed while maintaining original order.

cities_raw = ['Delhi', 'Mumbai', 'Delhi', 'Bangalore', 'Mumbai', 'Chennai', 'Delhi']

cleaned_cities = []
for a in cities_raw:
    if a not in cleaned_cities:
        cleaned_cities.append(a)
print(cleaned_cities)

# Expected Output:
# ['Delhi', 'Mumbai', 'Bangalore', 'Chennai']
```

```python
# Exercise 10: Word Frequency Counter

# Problem:
# Given a sentence,
# count the frequency of each word and return it as a dictionary.

sentence = 'data analyst uses data to find patterns in data every day'

words = sentence.split()
frequency = {}
for w in words:
    if w not in frequency:
        frequency[w] = 1
    else:
        frequency[w] += 1
print(frequency)

# Expected Output:
# {
#   'data': 3,
#   'analyst': 1,
#   'uses': 1,
#   'to': 1,
#   'find': 1,
#   'patterns': 1,
#   'in': 1,
#   'every': 1,
#   'day': 1
# }

# hint lena pda, give up kr diya tha maine
```
