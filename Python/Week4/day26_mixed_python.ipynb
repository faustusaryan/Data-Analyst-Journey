```python
# Q1. Given the employees list, write code that counts how many employees
# belong to each department.
# Output: {'IT': 4, 'HR': 2, 'Finance': 2, 'Marketing': 2}

employees = [
    {'id':1, 'name':'Rajesh',  'dept':'IT',        'salary':75000, 'city':'Delhi'},
    {'id':2, 'name':'Sunita',  'dept':'HR',        'salary':55000, 'city':'Mumbai'},
    {'id':3, 'name':'Aakash',  'dept':'IT',        'salary':60000, 'city':'Delhi'},
    {'id':4, 'name':'Meena',   'dept':'Finance',   'salary':80000, 'city':'Ahmedabad'},
    {'id':5, 'name':'Deepak',  'dept':'IT',        'salary':45000, 'city':'Bangalore'},
    {'id':6, 'name':'Prerna',  'dept':'HR',        'salary':42000, 'city':'Delhi'},
    {'id':7, 'name':'Ankit',   'dept':'Marketing', 'salary':50000, 'city':'Mumbai'},
    {'id':8, 'name':'Kavita',  'dept':'Finance',   'salary':72000, 'city':'Pune'},
    {'id':9, 'name':'Suresh',  'dept':'Marketing', 'salary':38000, 'city':'Chennai'},
    {'id':10,'name':'Pooja',   'dept':'IT',        'salary':68000, 'city':'Noida'},
  ]

dept_count = {}
for a in employees:
    dept = a['dept']
    if dept not in dept_count:
        dept_count[dept] = 1
    else:
        dept_count[dept] += 1
print(dept_count)
```

```python
# Q2. Calculate the average salary for each department.
# Round to 0 decimal places.
# Output: {'IT': 62000.0, 'HR': 48500.0, 'Finance': 76000.0, 'Marketing': 44000.0}

employees = [
    {'id':1, 'name':'Rajesh',  'dept':'IT',        'salary':75000, 'city':'Delhi'},
    {'id':2, 'name':'Sunita',  'dept':'HR',        'salary':55000, 'city':'Mumbai'},
    {'id':3, 'name':'Aakash',  'dept':'IT',        'salary':60000, 'city':'Delhi'},
    {'id':4, 'name':'Meena',   'dept':'Finance',   'salary':80000, 'city':'Ahmedabad'},
    {'id':5, 'name':'Deepak',  'dept':'IT',        'salary':45000, 'city':'Bangalore'},
    {'id':6, 'name':'Prerna',  'dept':'HR',        'salary':42000, 'city':'Delhi'},
    {'id':7, 'name':'Ankit',   'dept':'Marketing', 'salary':50000, 'city':'Mumbai'},
    {'id':8, 'name':'Kavita',  'dept':'Finance',   'salary':72000, 'city':'Pune'},
    {'id':9, 'name':'Suresh',  'dept':'Marketing', 'salary':38000, 'city':'Chennai'},
    {'id':10,'name':'Pooja',   'dept':'IT',        'salary':68000, 'city':'Noida'},
  ]

sum_dept_sal = {}
count = {}
for a in employees:
    dept = a['dept']
    salary = a['salary']
    if dept not in sum_dept_sal:
        sum_dept_sal[dept] = salary
        count[dept] = 1
    else:
        sum_dept_sal[dept] += salary
        count[dept] += 1

avg_sal = {}
for a in sum_dept_sal:
    avg_sal[a] = round(sum_dept_sal[a] / count[a], 0)
print(avg_sal)
```

```python
# Q3. Find the maximum salary in each department.
# Output: {'IT': 75000, 'HR': 55000, 'Finance': 80000, 'Marketing': 50000}

employees = [
    {'id':1, 'name':'Rajesh',  'dept':'IT',        'salary':75000, 'city':'Delhi'},
    {'id':2, 'name':'Sunita',  'dept':'HR',        'salary':55000, 'city':'Mumbai'},
    {'id':3, 'name':'Aakash',  'dept':'IT',        'salary':60000, 'city':'Delhi'},
    {'id':4, 'name':'Meena',   'dept':'Finance',   'salary':80000, 'city':'Ahmedabad'},
    {'id':5, 'name':'Deepak',  'dept':'IT',        'salary':45000, 'city':'Bangalore'},
    {'id':6, 'name':'Prerna',  'dept':'HR',        'salary':42000, 'city':'Delhi'},
    {'id':7, 'name':'Ankit',   'dept':'Marketing', 'salary':50000, 'city':'Mumbai'},
    {'id':8, 'name':'Kavita',  'dept':'Finance',   'salary':72000, 'city':'Pune'},
    {'id':9, 'name':'Suresh',  'dept':'Marketing', 'salary':38000, 'city':'Chennai'},
    {'id':10,'name':'Pooja',   'dept':'IT',        'salary':68000, 'city':'Noida'},
  ]

max_sal = {}
for a in employees:
    dept = a['dept']
    salary = a['salary']
    if dept not in max_sal:
        max_sal[dept] = salary
    else:
        if salary > max_sal[dept]:   
            max_sal[dept] = salary
print(max_sal)
```

```python
# Q4. Filter employees who earn strictly more than 60000
# AND belong to IT or Finance department.
# Return a list of their names.
# Output: ['Rajesh', 'Meena', 'Kavita', 'Pooja']
# Note: Aakash (IT, 60000) is NOT included — condition is strictly > 60000.

employees = [
    {'id':1, 'name':'Rajesh',  'dept':'IT',        'salary':75000, 'city':'Delhi'},
    {'id':2, 'name':'Sunita',  'dept':'HR',        'salary':55000, 'city':'Mumbai'},
    {'id':3, 'name':'Aakash',  'dept':'IT',        'salary':60000, 'city':'Delhi'},
    {'id':4, 'name':'Meena',   'dept':'Finance',   'salary':80000, 'city':'Ahmedabad'},
    {'id':5, 'name':'Deepak',  'dept':'IT',        'salary':45000, 'city':'Bangalore'},
    {'id':6, 'name':'Prerna',  'dept':'HR',        'salary':42000, 'city':'Delhi'},
    {'id':7, 'name':'Ankit',   'dept':'Marketing', 'salary':50000, 'city':'Mumbai'},
    {'id':8, 'name':'Kavita',  'dept':'Finance',   'salary':72000, 'city':'Pune'},
    {'id':9, 'name':'Suresh',  'dept':'Marketing', 'salary':38000, 'city':'Chennai'},
    {'id':10,'name':'Pooja',   'dept':'IT',        'salary':68000, 'city':'Noida'},
  ]

emp_lst = []
for a in employees:
    if a['salary'] > 60000 and a['dept'] in ('IT', 'Finance'):
        emp_lst.append(a['name'])
print(emp_lst)
```

```python
# Q5. Add a 'band' key to each employee dict in the list.
# Rule: salary >= 70000 → 'Senior'  |  >= 50000 → 'Mid'  |  else → 'Junior'
# Modify the original list (no need to create a new list).
# Expected bands: Rajesh→Senior, Sunita→Mid, Aakash→Mid, Meena→Senior,
#   Deepak→Junior, Prerna→Junior, Ankit→Mid, Kavita→Senior, Suresh→Junior, Pooja→Mid

employees = [
    {'id':1, 'name':'Rajesh',  'dept':'IT',        'salary':75000, 'city':'Delhi'},
    {'id':2, 'name':'Sunita',  'dept':'HR',        'salary':55000, 'city':'Mumbai'},
    {'id':3, 'name':'Aakash',  'dept':'IT',        'salary':60000, 'city':'Delhi'},
    {'id':4, 'name':'Meena',   'dept':'Finance',   'salary':80000, 'city':'Ahmedabad'},
    {'id':5, 'name':'Deepak',  'dept':'IT',        'salary':45000, 'city':'Bangalore'},
    {'id':6, 'name':'Prerna',  'dept':'HR',        'salary':42000, 'city':'Delhi'},
    {'id':7, 'name':'Ankit',   'dept':'Marketing', 'salary':50000, 'city':'Mumbai'},
    {'id':8, 'name':'Kavita',  'dept':'Finance',   'salary':72000, 'city':'Pune'},
    {'id':9, 'name':'Suresh',  'dept':'Marketing', 'salary':38000, 'city':'Chennai'},
    {'id':10,'name':'Pooja',   'dept':'IT',        'salary':68000, 'city':'Noida'},
  ]

for a in employees:
    salary = a['salary']
    if salary >= 70000:
        a['band'] = 'Senior'
    elif salary >= 50000:
        a['band'] = 'Mid'
    else:
        a['band'] = 'Junior'
        
print(employees)
```

```python
# Q6. Group employees by (department, city) combination.
# Key = tuple (dept, city), Value = list of employee names.
# Output (sample): {('IT','Delhi'): ['Rajesh','Aakash'], ('IT','Bangalore'): ['Deepak'], ...}

employees = [
    {'id':1, 'name':'Rajesh',  'dept':'IT',        'salary':75000, 'city':'Delhi'},
    {'id':2, 'name':'Sunita',  'dept':'HR',        'salary':55000, 'city':'Mumbai'},
    {'id':3, 'name':'Aakash',  'dept':'IT',        'salary':60000, 'city':'Delhi'},
    {'id':4, 'name':'Meena',   'dept':'Finance',   'salary':80000, 'city':'Ahmedabad'},
    {'id':5, 'name':'Deepak',  'dept':'IT',        'salary':45000, 'city':'Bangalore'},
    {'id':6, 'name':'Prerna',  'dept':'HR',        'salary':42000, 'city':'Delhi'},
    {'id':7, 'name':'Ankit',   'dept':'Marketing', 'salary':50000, 'city':'Mumbai'},
    {'id':8, 'name':'Kavita',  'dept':'Finance',   'salary':72000, 'city':'Pune'},
    {'id':9, 'name':'Suresh',  'dept':'Marketing', 'salary':38000, 'city':'Chennai'},
    {'id':10,'name':'Pooja',   'dept':'IT',        'salary':68000, 'city':'Noida'},
  ]

grouped  = {}

for a in employees:
    group = (a['dept'], a['city'])
    name = a['name']
    if group not in grouped :
        grouped [group] = [name]
    else:
        grouped [group].append(name)
print(grouped )
```

```python
# Q9. Create a dict where key = department, value = list of employee names in that dept.
# Output: {'IT': ['Rajesh','Aakash','Deepak','Pooja'],
#          'HR': ['Sunita','Prerna'],
#          'Finance': ['Meena','Kavita'],
#          'Marketing': ['Ankit','Suresh']}

employees = [
    {'id':1, 'name':'Rajesh',  'dept':'IT',        'salary':75000, 'city':'Delhi'},
    {'id':2, 'name':'Sunita',  'dept':'HR',        'salary':55000, 'city':'Mumbai'},
    {'id':3, 'name':'Aakash',  'dept':'IT',        'salary':60000, 'city':'Delhi'},
    {'id':4, 'name':'Meena',   'dept':'Finance',   'salary':80000, 'city':'Ahmedabad'},
    {'id':5, 'name':'Deepak',  'dept':'IT',        'salary':45000, 'city':'Bangalore'},
    {'id':6, 'name':'Prerna',  'dept':'HR',        'salary':42000, 'city':'Delhi'},
    {'id':7, 'name':'Ankit',   'dept':'Marketing', 'salary':50000, 'city':'Mumbai'},
    {'id':8, 'name':'Kavita',  'dept':'Finance',   'salary':72000, 'city':'Pune'},
    {'id':9, 'name':'Suresh',  'dept':'Marketing', 'salary':38000, 'city':'Chennai'},
    {'id':10,'name':'Pooja',   'dept':'IT',        'salary':68000, 'city':'Noida'},
  ]

dept_emp = {}
for a in employees:
    dept = a['dept']
    emp = a['name']
    if dept not in dept_emp:
        dept_emp[dept] = [emp]
    else:
        dept_emp[dept].append(emp)

print(dept_emp)
```
