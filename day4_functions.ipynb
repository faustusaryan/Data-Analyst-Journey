```python
# Task 1:
# def salary_grade(sal) banao

# 45000 se zyada → "High"
# 35000 se zyada → "Medium"
# Baaki → "Low"

# Test: salary_grade(52000), salary_grade(38000), salary_grade(30000)
```

```python
def salary_grade(slr):
    if slr >= 45000:
        return "High"
    elif slr >= 35000 and slr < 45000:
        return "Medium"
    else:
        return"Low"

print(salary_grade(52000))
print(salary_grade(38000))
print(salary_grade(30000))
```

```python
# Task 2:
# def calc_tax(salary) banao

# 40000 se zyada → 10% tax
# 40000 ya kam → 5% tax
# Tax amount return karo (salary nahi)

# Test: calc_tax(45000), calc_tax(35000)
```

```python
def calc_tax(slr):
    if slr >= 40000:
        return slr*.10
    elif slr < 40000:
        return slr*.05

print(calc_tax(45000))
print(calc_tax(35000))
```

```python
# Task 3:
# def dept_avg(emp_list) banao

# Input: list of dicts (har dict mein name, dept, salary keys hain)
# Output: {dept: avg_salary} dict

# Test data:
# pythonemps = [
#     {"name":"Raj", "dept":"IT", "salary":45000},
#     {"name":"Amit", "dept":"IT", "salary":52000},
#     {"name":"Priya", "dept":"HR", "salary":38000},
#     {"name":"Vivek", "dept":"HR", "salary":35000}
# ]
```

```python
def dept_avg(lst):
    count = {}
    total = {}
    for emp in lst:
        total[emp["dept"]] = total.get(emp["dept"], 0) + emp["salary"]
        count[emp["dept"]] = count.get(emp["dept"], 0) + 1

    result = {}
    for dept in total:
        result[dept] = total[dept] / count[dept]
    return result 
   

c = [
    {"name":"Raj", "dept":"IT", "salary":45000},
    {"name":"Amit", "dept":"IT", "salary":52000},
    {"name":"Priya", "dept":"HR", "salary":38000},
    {"name":"Vivek", "dept":"HR", "salary":35000}
]

dept_avg(c)

### dimag kharab ho gaya yrr it is tough, nhi hua khudse help leni pdi

    
```
