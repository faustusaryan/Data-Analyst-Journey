```python
# Create a dict with 3 student names and marks. Print only those with marks > 60

dict_0 = {
    "Aryan" : 59,
    "Gauri" : 97,
    "Vishal" : 88,
}
for name, marks in dict_0.items():
    if marks > 60:
        print(name, "=", marks)
```

```python
# Write a function that checks if a number is even or odd

def check_odd_even(num):
    if num % 2 == 0:
        print(num, "is even")
    else:
        print(num, "is odd")

check_odd_even(62)
```

```python
# Reverse a list [1, 2, 3, 4, 5] WITHOUT using .reverse()

lst = [1, 2, 3, 4, 5]
new_lst = []
i = len(lst) - 1

while i >= 0:
    new_lst.append(lst[i])
    i -= 1
print(new_lst)
```
