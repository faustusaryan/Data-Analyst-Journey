```python
# F1
# Question:
# greet(name)
# Naam leke "Hello [name]!" return karo.

# Hint:
# - def + return
# - string concat
# - "Hello " + name + "!"

def greet(name):
    return 'hello ' + name + '!'
greet('Aryan')
```

```python
# F2
# Question:
# calc_bonus(salary)
# Salary ka 10% return karo.

# Hint:
# - salary * 0.10
# - result float aayega

def calc_bonus(salary):
    return salary*0.1
calc_bonus(45000)
```

```python
# F3
# Question:
# salary_grade(salary)

# 45000 ya usse zyada → 'Senior'
# 35000–44999 → 'Mid'
# 35000 se neeche → 'Junior'

# Hint:
# - if / elif / else
# - sabse badi condition pehle check karo

def salary_grade(salary):
    if salary >= 45000:
        return 'Senior'
    elif salary >= 35000 and salary <= 44999:
        return 'Mid'
    else:
        return 'Junior'

salary_grade(44000)
```

```python
# F4
# Question:
# is_it_dept(dept)

# Agar dept == "IT" → True
# warna → False

def is_it_dept(dept):
    if dept == 'IT':
        return True
    else:
        return False

is_it_dept('HR')
```

```python
# F5
# Question:
# annual_salary(monthly)

# Monthly salary ka 12 mahine ka total return karo.

def annual_salary(monthly_salary):
    return monthly_salary*12

annual_salary(10000)
```

```python
# F6
# Question:
# tax_amount(salary, rate=0.1)

# Default tax rate 10%.
# Salary * rate return karo.

def tax_amount(salary, rate = 0.1):
    return salary*rate

tax_amount(45000)
```

```python
# F7
# Question:
# net_salary(salary, rate=0.1)

# Salary mein se tax_amount(salary, rate) minus karo.

def tax_amount(salary, rate = 0.1):
    return salary*rate
    
def net_salary(salary, rate = 0.1):
    return salary - tax_amount(salary, rate)

net_salary(65000, 0.15)
    
```

```python
# F8
# Question:
# emp_info(name, dept)

# Name uppercase
# Dept lowercase
# Tuple mein dono return karo.

def emp_info(name, dept):
    return (name.upper() , dept.lower())
    
emp_info('Aryan', 'AI')
```

```python
# F9
# Question:
# city_check(city, target_city="Delhi")

# Agar city == target_city → "Local"
# warna → "Outstation"

def city_check(city, target_city = 'Delhi'):
    if city == target_city:
        return 'Local'
    else:
        return 'Outstation'

city_check('Lucknow')
```

```python
# F10
# Question:
# count_depts(dept_list)

# List mein har department kitni baar aya hai count karo.
# Result dictionary mein return karo.

def count_depts(dept_list):
    dept_count = {}
    for dept in dept_list:
        dept_count[dept] = dept_count.get(dept, 0) + 1
    return dept_count
a = count_depts(['IT', 'HR', 'AI', 'IT', 'HR'])
print(a)

#failed
```

```python
# F11
# Question:
# filter_high_earners(salary_list, cutoff=40000)

# Sirf wo salaries return karo jo cutoff se strictly zyada ho.

def filter_high_earners(salary_list, cut_off = 40000):
    return [high for high in salary_list if high > cut_off]

print(filter_high_earners([35000, 42000, 29000, 50000, 80000, 40000]))
```

```python
# F12
# Question:
# total_salary(*salaries)

# Kitni bhi salaries aayein — sabka total return karo.

def total_salary(*salaries):
    return sum(salaries)

total_salary(1000, 2000, 3000, 4000, 5000)
```

```python
# F13
# Question:
# dept_budget_check(dept, salary, budget_dict)

# budget_dict mein dept ka budget hoga.
# Agar salary < budget → "Under Budget"
# warna → "Over Budget

def dept_budget_check(dept, salary, budget_dict):
    if salary < budget_dict[dept]:
        return 'Under Budget'
    else:
        return 'Over Budget'

dept_budget_check('IT', 45000, {'IT' : 58000})
```

```python
# F14
# Question:
# name_format(first, last="")

# Agar last diya ho → "First Last"
# warna → sirf "First"

def name_format(first, last = ''):
    if last != '':
        return first + ' ' + last
    else:
        return first

name_format('Aryan')
```

```python
# F15
# Question:
# summarize_emp(name, salary, dept)

# Readable string return karo:
# "[name] works in [dept] and earns Rs.[salary]/month"

def summarize_emp(name, salary, dept):
    return name + ' works in ' + dept + ' and earns Rs.' + str(salary) + '/month'

summarize_emp('Aryan', 58000, 'AI')

# and f-string use krna nhi aata
```

```python
# B1
# Question:
# is_senior(salary, cutoff=45000)

# Agar salary >= cutoff ho → True
# warna → False return karo.

def is_senior(salary, cut_off = 45000):
    if salary >= cut_off:
        return True
    else:
        return False

is_senior(46000)
```

```python
# B2
# Question:
# get_dept_employees(emp_list, dept)

# Pehle ek list banao:

# emp_list = [
# {'name':'Raj','dept':'IT'},
# {'name':'Priya','dept':'HR'},
# {'name':'Amit','dept':'IT'}
# ]

# Function ka kaam:
# sirf us dept ke employees return karo.

def get_dept_employees(emp_lst, dept):
    return [emp for emp in emp_lst if emp['dept'] == dept]

get_dept_employees([
{'name':'Raj','dept':'IT'},
{'name':'Priya','dept':'HR'},
{'name':'Amit','dept':'IT'}
], 'IT')

# a small help 
```

```python
# B3
# Question:
# apply_raise(salary, percent)

# Salary mein percent% raise add karke new salary return karo.

# Example:
# apply_raise(45000, 10)

def apply_raise(salary, percent):
    return salary + salary * 1/100 * percent
apply_raise(45000, 10)
```
