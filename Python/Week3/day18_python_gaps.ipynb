```python
# P01:- Count Employees per Department

employees = [
    {"name": "Aarav",  "dept": "Engineering", "salary": 75000, "gender": "M", "years": 3},
    {"name": "Priya",  "dept": "Marketing",   "salary": 55000, "gender": "F", "years": 2},
    {"name": "Rahul",  "dept": "Engineering", "salary": 82000, "gender": "M", "years": 5},
    {"name": "Sneha",  "dept": "HR",          "salary": 48000, "gender": "F", "years": 1},
    {"name": "Vikram", "dept": "Marketing",   "salary": 62000, "gender": "M", "years": 4},
    {"name": "Divya",  "dept": "Engineering", "salary": 70000, "gender": "F", "years": 2},
    {"name": "Karan",  "dept": "HR",          "salary": 50000, "gender": "M", "years": 3},
    {"name": "Neha",   "dept": "Marketing",   "salary": 58000, "gender": "F", "years": 1},
]
# Task:- employees list se ek dictionary banao jisme key = department name, value = us department mein kitne employees hain.
# Expected Output:- {'Engineering': 3, 'Marketing': 3, 'HR': 2}

dept_count = {}
for el in employees:
    dept = el['dept']
    if dept in dept_count:
        dept_count[dept] +=1
    else:
        dept_count[dept] = 1
print(dept_count)
```

```python
# P02:- Total Salary Bill per Department
employees = [
    {"name": "Aarav",  "dept": "Engineering", "salary": 75000, "gender": "M", "years": 3},
    {"name": "Priya",  "dept": "Marketing",   "salary": 55000, "gender": "F", "years": 2},
    {"name": "Rahul",  "dept": "Engineering", "salary": 82000, "gender": "M", "years": 5},
    {"name": "Sneha",  "dept": "HR",          "salary": 48000, "gender": "F", "years": 1},
    {"name": "Vikram", "dept": "Marketing",   "salary": 62000, "gender": "M", "years": 4},
    {"name": "Divya",  "dept": "Engineering", "salary": 70000, "gender": "F", "years": 2},
    {"name": "Karan",  "dept": "HR",          "salary": 50000, "gender": "M", "years": 3},
    {"name": "Neha",   "dept": "Marketing",   "salary": 58000, "gender": "F", "years": 1},
]
# Task:- Har department ka total salary sum karo. Output: dict with dept → total salary.
# Expected Output:- {'Engineering': 227000, 'Marketing': 175000, 'HR': 98000}

total_sal = {}
for emp in employees:
    dept = emp['dept']
    sal = emp['salary']
    if dept in total_sal:
        total_sal[dept] += sal
    else:
        total_sal[dept] = sal
print(total_sal)    

# Alternative:

dept_total = {}
for emp in employees:
    dept = emp['dept']
    dept_total[dept] = dept_total.get(dept, 0) + emp['salary']
print(dept_total)
```

```python
# P03:- Average Salary per Department
employees = [
    {"name": "Aarav",  "dept": "Engineering", "salary": 75000, "gender": "M", "years": 3},
    {"name": "Priya",  "dept": "Marketing",   "salary": 55000, "gender": "F", "years": 2},
    {"name": "Rahul",  "dept": "Engineering", "salary": 82000, "gender": "M", "years": 5},
    {"name": "Sneha",  "dept": "HR",          "salary": 48000, "gender": "F", "years": 1},
    {"name": "Vikram", "dept": "Marketing",   "salary": 62000, "gender": "M", "years": 4},
    {"name": "Divya",  "dept": "Engineering", "salary": 70000, "gender": "F", "years": 2},
    {"name": "Karan",  "dept": "HR",          "salary": 50000, "gender": "M", "years": 3},
    {"name": "Neha",   "dept": "Marketing",   "salary": 58000, "gender": "F", "years": 1},
]
# Task:- Har department ka average salary calculate karo. Result round karke 2 decimal places mein do.
# Expected Output:- {'Engineering': 75666.67, 'Marketing': 58333.33, 'HR': 49000.0}

dept_sum = {}
dept_count = {}
for empl in employees:
    dept = empl['dept']
    sal = empl['salary']
    if dept in dept_sum:
        dept_sum[dept] += sal
        dept_count[dept] += 1
    else:
        dept_sum[dept] = sal
        dept_count[dept] = 1
dept_avg = {}
for dept in dept_sum:
    dept_avg[dept] = round(dept_sum[dept]/dept_count[dept], 2)
print(dept_avg)

# Alternative:

dept_total = {}
dept_count = {}
 
for emp in employees:
    dept = emp['dept']
    dept_total[dept] = dept_total.get(dept, 0) + emp['salary']
    dept_count[dept] = dept_count.get(dept, 0) + 1
 
dept_avg = {}
for dept in dept_total:
    dept_avg[dept] = round(dept_total[dept] / dept_count[dept], 2)
    
print(dept_avg)
```
