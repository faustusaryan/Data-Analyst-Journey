```python
# L1-Q1 · Dictionary basics — key access
# Ek employee dict hai. Uski salary aur dept print karo.
emp = {"name": "Aarav", "dept": "Engineering", "salary": 75000}

print(emp['salary'], emp['dept'])
```

```python
# L1-Q2 · dict.get() — the most important pattern
# d.get(key, default) — key nahi hai toh error nahi aata, default return hota hai.
d = {"Engineering": 5, "HR": 2}

val1 = d.get('IT', 'not found')
val2 = d.get('HR', 0)
print(val1)
print(val2)
```

```python
# L1-Q3 · dict.get() — counting pattern
# Ye P01 ka exact engine hai. Ek word list hai, har word kitni baar aaya count karo.
words = ["apple","banana","apple","cherry","banana","apple"]

count_word = {}
for word in words:
    if word not in count_word:
        count_word[word] = 1
    else:
        count_word[word] += 1
print(count_word)

#Alternative

word_count = {}
for w in words:
    word_count[w] = count_word.get(w, 0) + 1
print(word_count)        
```

```python
# L1-Q4 · dict.get() — summing pattern
# Ye P02 ka exact engine hai. Har city ka total score sum karo.
data = [
    {"city": "Delhi",  "score": 10},
    {"city": "Mumbai", "score": 20},
    {"city": "Delhi",  "score": 15},
    {"city": "Mumbai", "score": 5},
    {"city": "Delhi",  "score": 8},
]

total_score = {}
for t in data:
    city = t['city']
    total_score[city] = total_score.get(city, 0) + t['score']
print(total_score)
```
