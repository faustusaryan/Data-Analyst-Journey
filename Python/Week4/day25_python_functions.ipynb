```python
# Q1. Given a list of integers that may contain duplicate values, implement a function
#     get_unique(lst) that returns a new list with only the unique elements,
#     preserving the order in which they first appeared in the input.
 
#     Input:  [1, 2, 2, 3, 1, 4, 3, 5]
#     Output: [1, 2, 3, 4, 5]
 
#     Constraint: Do not use Python's built-in set() — maintain order manually.

def get_unique(num_list):
    lst = []
    for a in num_list:
        if a not in lst:
            lst.append(a)
    return lst

get_unique([1, 2, 2, 3, 1, 4, 3, 5])
```

```python
# Q2. Write a function sort_by_key(data, key, reverse=False) that accepts a list of
#     dictionaries and sorts them by the value of a specified key field.
#     Default order should be ascending; pass reverse=True for descending.
 
#     Input:  [{'name':'Rajesh','salary':75000}, {'name':'Deepak','salary':45000}, ...]
#     Output (ascending by salary): Deepak(45k) -> Rajesh(75k) -> Meena(80k)

def sort_by_key(data, key, reverse=False):
    return sorted(data, key = lambda x: x[key], reverse = reverse)

x = [{'name':'Rajesh','salary':75000}, 
     {'name':'Deepak','salary':45000},
     {'name':'Meena','salary':80000}]
a = sort_by_key(x, 'salary')
b = sort_by_key(x, 'salary', reverse=True)
print(a)
print(b)
```

```python
# Q3. Given a list of numbers, write a function calculate_stats(numbers) that
#     computes and returns a dictionary with three statistical values:
 
#       mean   = sum / count
#       median = middle value of sorted list
#                (average of two middle values if count is even)
#       mode   = value that appears most frequently
 
#     Input:  [75000, 55000, 60000, 80000, 45000, 60000, 50000]
#     Output: {'mean': 60714.29, 'median': 60000, 'mode': 60000}

def calculate_stats(numbers):
    mean = round(sum(numbers)/len(numbers), 2)
    
    sorted_nums = sorted(numbers)  
    
    if len(sorted_nums) % 2 == 0:
        median = (sorted_nums[len(sorted_nums)//2] + sorted_nums[len(sorted_nums)//2 - 1]) / 2
    else:
        median = sorted_nums[len(sorted_nums)//2]  
        
    dict2 = {}
    for a in numbers:
        if a not in dict2:
            dict2[a] = 1
        else:
            dict2[a] += 1
    mode = max(dict2, key=dict2.get)
    
    return {'mean': mean, 'median': median, 'mode': mode}

calculate_stats([75000, 55000, 60000, 80000, 45000, 60000, 50000])
```

```python
# Q4. You are given a list of tuples of the form (product_name, price).
#     Using a lambda function, sort this list by price in two ways:
#       sorted_asc  : ascending  (low price first)
#       sorted_desc : descending (high price first)
 
    # Input:  [('Laptop',55000), ('Mouse',500), ('Headphones',3500), ('Monitor',18000)]
    # Output (asc): Mouse -> Headphones -> Monitor -> Laptop

xyz = [('Laptop',55000), ('Mouse',500), ('Headphones',3500), ('Monitor',18000)]

sorted_asc = sorted(xyz, key=lambda x:x[1])
sorted_desc = sorted(xyz, key=lambda x:x[1], reverse=True)

print(sorted_asc, '\n', sorted_desc)
```

```python
# Q5. Write a function calculate_average(*scores) that accepts any number of
#     numeric arguments and returns their average as a float.
#     The function must work regardless of how many values are passed.
 
#     calculate_average(85, 92, 78)          ->  85.0
#     calculate_average(70, 88, 95, 67, 80)  ->  80.0

def calculate_avg(*scores):
    return sum(scores)/len(scores)

calculate_avg(70, 88, 95, 67, 80)    
```

```python
# Q6. Implement a function employee_details(**kwargs) that accepts any number of
#     keyword arguments representing employee attributes.
#     Print each attribute on a separate line in the format:  key: value
 
#     Input:  employee_details(name='Rajesh', dept='IT', salary=75000, city='Delhi')
#     Output:
#       name: Rajesh
#       dept: IT
#       salary: 75000
#       city: Delhi

def emp_detail(**key_word):
    for key, value in key_word.items():
        print(key, ':', value)  

emp_detail(name='Rajesh', dept='IT', salary=75000, city='Delhi')
```

```python
# Q7. Given a string, write a function most_frequent_char(text) that returns
#     the character appearing the most number of times.
#     Ignore whitespace (spaces) while counting.
#     If there is a tie, returning any one of the top characters is acceptable.
 
#     Input:  'data analyst'
#     Output: 'a'   (appears 4 times)

def most_freq_char(text):
    text_lst = list(text)
    count = {}
    for a in text_lst:
        if a not in count:
            count[a] = 1
        else:
            count[a] += 1
    return max(count, key=count.get)

most_freq_char('data analyst')
```

```python
# Q8. You are given a dictionary where keys are month numbers (1-12) and
#     values are the revenue for that month.
#     Write a function quarterly_summary(monthly_data) that returns a dictionary
#     with total revenue grouped into four quarters:
 
#     Q1 = Jan-Mar (1,2,3)   Q2 = Apr-Jun (4,5,6)
#     Q3 = Jul-Sep (7,8,9)   Q4 = Oct-Dec (10,11,12)
 
#     Output: {'Q1': 155000, 'Q2': 205000, 'Q3': 205000, 'Q4': 270000}

def quarterly_summary(monthly_data):
    Q1 = 0
    Q2 = 0
    Q3 = 0
    Q4 = 0
    for a, b in monthly_data.items():
        if a < 4:
            Q1 += b
        elif a < 7 and a > 3:
            Q2 += b
        elif a < 10 and a > 6:
            Q3 += b
        elif a > 9:
            Q4 += b
    return {'Q1': Q1, 'Q2': Q2, 'Q3': Q3, 'Q4': Q4}

quarterly_summary({
    1: 45000,  2: 55000,  3: 55000,   # Q1 = 155000
    4: 65000,  5: 70000,  6: 70000,   # Q2 = 205000
    7: 65000,  8: 70000,  9: 70000,   # Q3 = 205000
    10: 80000, 11: 90000, 12: 100000  # Q4 = 270000
})
```

```python
# Q9. Write a function validate_employee(emp) that validates an employee dictionary
#     against three rules. Return True if all pass, False with a printed reason if not.
 
#     Rule 1: emp['name'] must be a non-empty string.
#     Rule 2: emp['salary'] must be greater than 0.
#     Rule 3: emp['dept'] must be in: ['IT', 'HR', 'Finance', 'Marketing']
 
#     {'name':'Rajesh','salary':75000,'dept':'IT'}    ->  True
#     {'name':'',      'salary':75000,'dept':'IT'}    ->  False (empty name)
#     {'name':'Rajesh','salary':-1000,'dept':'IT'}    ->  False (invalid salary)
#     {'name':'Rajesh','salary':75000,'dept':'Legal'} ->  False (invalid dept)

def validate_emp(emp):
    dept = ['IT', 'HR', 'Finance', 'Marketing']
    
    if len(emp['name']) == 0:
        print('Invalid: empty name')
        return False
    if emp['salary'] <= 0:
        print('Invalid: invalid salary')
        return False
    if emp['dept'] not in dept:
        print('Invalid: invalid dept')
        return False
    
    return True

print(validate_emp({'name':'Rajesh','salary':75000,'dept':'IT'}))     
print(validate_emp({'name':'',      'salary':75000,'dept':'IT'}))     
print(validate_emp({'name':'Rajesh','salary':-1000,'dept':'IT'}))     
print(validate_emp({'name':'Rajesh','salary':75000,'dept':'Legal'}))  
```

```python
# Q10. You are given a list of 6 transaction dictionaries, each having fields:
#      id, customer, amount, and status (Delivered / Cancelled / Pending).
#      Solve the following three sub-problems:
#      Expected:  a) 63800   b) 11033.33   c) 'Amit'

transactions = [
    {'id':1,'customer':'Amit',  'amount':55500,'status':'Delivered'},
    {'id':2,'customer':'Priya', 'amount':3500, 'status':'Delivered'},
    {'id':3,'customer':'Rahul', 'amount':800,  'status':'Cancelled'},
    {'id':4,'customer':'Sneha', 'amount':1600, 'status':'Pending'},
    {'id':5,'customer':'Vikram','amount':500,  'status':'Delivered'},
    {'id':6,'customer':'Kavya', 'amount':4300, 'status':'Delivered'},
]

# (a) Find the total amount of all 'Delivered' transactions.

total_delv_amount = 0
for a in transactions:
    if a['status'] == 'Delivered':
        total_delv_amount += a['amount']
print(total_delv_amount)
    
# (b) Find the average amount across all 6 transactions.

total_amount = 0
for a in transactions:
    total_amount += a['amount']
avg_transc_amount = round(total_amount/len(transactions), 2)
print(avg_transc_amount)

# (c) Find the customer name with the single largest transaction amount.

max_transc = max(transactions, key=lambda x: x['amount'])
print(max_transc['customer'])
```
