```python
# EXAMPLE:-

animals = ['lion', 'tiger', 'monkey', 'elephant', 'frog']
filtered_animals = [animal.title() for animal in animals]
print(filtered_animals)
```

```python
# 1 se 10 tak ke numbers ki list banao — sirf even numbers (List comprehension use karo — loop nahi)

even_lst = [num for num in range(1,11) if num%2 == 0]
print(even_lst)
```

```python
# 1 se 10 tak har number ka square nikalo — list mein [1, 4, 9, 16, 25...]

square_lst = [num**2 for num in range(1,11)]
print(square_lst)
```

```python
# words = ["hello", "world", "python", "data"] Sirf 5 se zyada characters wale words nikalo

words = ["hello", "world", "python", "data"]
new_words = [word for word in words if len(word) > 5]
print(new_words)
```

```python
 # salaries = [45000, 38000, 52000, 35000, 43000] Har salary pe 10% bonus add karo — nai list banao

salaries = [45000, 38000, 52000, 35000, 43000]
include_bonus = [bonus + bonus*.10 for bonus in salaries]
print(include_bonus)
```
