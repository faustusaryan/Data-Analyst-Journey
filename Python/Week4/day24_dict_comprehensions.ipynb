```python
# Exercise 1: Basic Dict Comprehension
numbers = [1, 2, 3, 4, 5]
# Task: {number: square} dict banao
# Expected: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

dict1 = {a: a**2 for a in numbers}
print(dict1)
```

```python
# Exercise 2: Filter Dict Using Comprehension
salary_dict = {'Rajesh': 75000, 'Sunita': 55000,
               'Aakash': 60000, 'Meena': 80000, 'Deepak': 45000}
# Task: Sirf 60000 se zyada salary wale employees rakhein
# Expected: {'Rajesh': 75000, 'Aakash': 60000, 'Meena': 80000}

dict2 = {a: b for a,b in salary_dict.items() if b >= 60000}
print(dict2)
```

```python
# Exercise 3: Nested Dict — Access Karna
company = {
    'IT':        {'employees': 5, 'avg_salary': 60000, 'head': 'Rajesh'},
    'HR':        {'employees': 3, 'avg_salary': 48333, 'head': 'Sunita'},
    'Finance':   {'employees': 3, 'avg_salary': 69000, 'head': 'Meena'},
    'Marketing': {'employees': 3, 'avg_salary': 44000, 'head': 'Ankit'}
}
# Task a: IT department ka avg_salary print karo

print(company['IT']['avg_salary'])

# Task b: Sab departments ke heads ki list banao

lst = [b['head'] for a,b in company.items()]
print(lst)

# Task c: Highest avg_salary wala department find karo
# Expected c: 'Finance'

highest_dept = max(company, key=lambda x: company[x]['avg_salary'])
print(highest_dept)
```

```python
# Exercise 4: dict.get() Method
employee = {'name': 'Rajesh', 'dept': 'IT', 'salary': 75000}
# Task: 'city' key ka value nikalo — agar nahi hai to 'Not Specified' return ho
# Use .get() method, NOT try/except
# Expected: 'Not Specified'

print(employee.get('city', 'Not Specified'))
```

```python
# Exercise 5: Merge Two Dicts
basic_info = {'name': 'Amit',  'city': 'Delhi', 'age': 28}
job_info   = {'dept': 'IT', 'salary': 65000, 'city': 'Noida'}  # city override hogi
# Task: Dono dicts merge karo — job_info ki values override karein where conflict
# Expected: {'name': 'Amit', 'city': 'Noida', 'age': 28, 'dept': 'IT', 'salary': 65000}

merged = {**basic_info, **job_info}
print(merged)
```

```python
# Exercise 6: List of Dicts → Single Dict
students = [
    {'name': 'Amit',  'score': 85},
    {'name': 'Priya', 'score': 92},
    {'name': 'Rahul', 'score': 78}
]
# Task: {name: score} format ka dict banao
# Expected: {'Amit': 85, 'Priya': 92, 'Rahul': 78}
new_dict = {}
for a in students:
    new_dict[a['name']] = a['score']
print(new_dict)

dict11 = {a['name']: a['score'] for a in students}
print(dict11)
```

```python
# Exercise 7: Sort Dict by Value
product_sales = {'Laptop': 15, 'Mouse': 45, 'Headphones': 22,
                 'Monitor': 8, 'Keyboard': 30}
# Task: Sales ke hisaab se descending order mein sort karo
# Expected: {'Mouse': 45, 'Keyboard': 30, 'Headphones': 22, 'Laptop': 15, 'Monitor': 8}

sorted_dict = dict(sorted(product_sales.items(), key = lambda x: x[1], reverse = True))
print(sorted_dict)
```

```python
# Exercise 8: Count Occurrences
statuses = ['Delivered', 'Cancelled', 'Delivered', 'Pending',
            'Delivered', 'Cancelled', 'Delivered']
# Task: Har status ki count wala dict banao
# Expected: {'Delivered': 4, 'Cancelled': 2, 'Pending': 1}

status_dict = {}
for a in statuses:
    if a not in status_dict:
        status_dict[a] = 1
    else:
        status_dict[a] +=1
print(status_dict)
```

```python
# Exercise 9: Assign Grades
marks = {'Amit': 85, 'Priya': 92, 'Rahul': 55, 'Sneha': 73, 'Vikram': 41}
# Grade rule: 90+ = 'A', 75-89 = 'B', 60-74 = 'C', below 60 = 'F'
# Task: {name: grade} dict banao
# Expected: {'Amit': 'B', 'Priya': 'A', 'Rahul': 'F', 'Sneha': 'C', 'Vikram': 'F'}

grade = {name: 'A' if score >= 90 else 'B' if score >= 75 else 'C' if score >= 60 else 'F'
    for name, score in marks.items()}
print(grade)
```

```python
# Exercise 10: .items(), .keys(), .values() Practice
dept_budgets = {'IT': 500000, 'HR': 200000, 'Finance': 800000, 'Marketing': 300000}
# Task a: Sabse zyada budget wale department ka naam aur budget print karo

max_budget = max(dept_budgets, key = lambda x:dept_budgets[x])
print(max_budget, dept_budgets[max_budget])    

# Task b: Total budget calculate karo using .values()

total_budget = sum(dept_budgets.values())
print(total_budget)

# Task c: Print all department names using .keys()

print(dept_budgets.keys())
```
