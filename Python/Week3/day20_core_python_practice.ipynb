### L1-Q5 · Loop over list of dicts
List of dicts mein loop karke specific fields print karna.
Har student ka naam aur marks print karo
# Expected:
* Ravi scored 85
* Sita scored 92
* Mohan scored 78

```python
students = [
    {"name": "Ravi", "marks": 85},
    {"name": "Sita", "marks": 92},
    {"name": "Mohan", "marks": 78},
]

for s in students:
    print(s['name'], 'scored', s['marks'])
```

### L1-Q6 · Define a simple function
### is_pass(marks) function likho
### 50 ya usse zyada = "Pass", warna "Fail"


```python
def is_pass(marks):
    if marks >= 50:
        return 'Pass'
    else: 
        return 'Fail'

print(is_pass(75))  
print(is_pass(40))  
print(is_pass(50))  
```

### L1-Q7 · Function with multiple conditions (slab logic)
* get_grade(marks) function likho:
* 90+      → "A"
* 75-89    → "B"
* 50-74    → "C"
* below 50 → "F"

```python
def get_grade(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 75:
        return 'B'
    elif marks >= 50:
        return 'C'
    else:
        return 'F'
        
print(get_grade(95)) 
print(get_grade(80))  
print(get_grade(60))  
print(get_grade(30))  
```

### L1-Q8 · sorted() + lambda
* price ke hisaab se ascending (low to high) sort karo
* price ke hisaab se descending (high to low) sort karo

```python
products = [
    {"name": "Phone", "price": 15000},
    {"name": "Laptop", "price": 55000},
    {"name": "Tablet", "price": 25000},
    {"name": "Watch", "price": 8000},
]
asc = sorted(products, key=lambda p: p['price'])
desc = sorted(products, key=lambda p: p['price'], reverse = True)
print('Ascending:-')
for p in asc:
    print(p['name'], p['price'])
print('Descending:-')
for p in desc:
    print(p['name'], p['price'])
```

### L1-Q9 · List comprehension — filter
* List comprehension se sirf 10 se bade numbers nikalo
* List comprehension se sirf even numbers nikalo

```python
numbers = [4, 17, 3, 22, 8, 31, 15, 6]
list1 = [n for n in numbers if n > 10]
print(list1)
list2 = [a for a in numbers if a%2 == 0]
print(list2)
```

### L1-Q10 · String methods: strip, lower, upper
* Spaces hatao (strip) phir lowercase karo
* Spaces hatao phir proper title case karo ("Engineering")

```python
raw = "  ENGINEERING  "
print(raw.strip().lower())
print(raw.strip().title())
```

### L1-Q11 · try/except — error handling
* Har value ko int mein convert karo, Agar error aaye toh 0 print karo

```python
values = ["100", "abc", None, "250", "xyz"]
for v in values:
    try:
        print(int(v))
    except:
        print(0)
```
