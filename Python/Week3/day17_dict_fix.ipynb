```python
# Simple dict banao — {name: 'Raj', salary: 45000, dept: 'IT'} — access karo

info = {'name': 'Raj', 'salary': 45000, 'dept': 'IT'}
print(info)
```

```python
# Function jo salary badhaye — 10% increment add karo dict mein

def sal_inc(emp):
    emp['salary'] = emp['salary'] * 1.1
    return round(emp['salary'], 2)
info = {'name': 'Raj', 'salary': 45000, 'dept': 'IT'}
print(sal_inc(info))
```

```python
# C1:- List of names aur list of marks hai. Dict banao jisme naam key ho aur marks value. 
# (Normal for-loop se pehle, phir dict comprehension se.)	

names = ['Aryan', 'Aman', 'Gauri', 'Faustus']
marks = [95, 49, 94, 99]

stu = {}
for i in range(len(names)):
    stu[names[i]] = marks[i]
print(stu)

student = {names[i]: marks[i] for i in range(len(names))}
print(student)
```

```python
# C2:- 1 se 10 tak ke numbers ka cube dict banao. Key = number, Value = number**3. Comprehension mein karo.

dict_cube = {i: i**3 for i in range(1,11)}
print(dict_cube)
```

```python
# C3:- Pehle ek scores dict banao: scores = {'Rahul':85, 'Priya':62, 'Amit':91, 'Sunita':78, 'Kavya':88}. 
# Phir is dict mein se sirf 80 se zyada score waale students ki filtered dict banao.

scores = {'Rahul':85, 'Priya':62, 'Amit':91, 'Sunita':78, 'Kavya':88}

new_dict = {}
for keys, vals in scores.items():
    if vals > 80:
        new_dict[keys] = vals
print(new_dict)

new_scores = {key: val for key, val in scores.items() if val > 80}
print(new_scores)
```

```python
# C11:- Employees ko department ke hisaab se group karo. 
# Har department mein kitne employees hain — print karo.
employees = [
    {'name': 'Rahul',  'dept': 'IT',      'salary': 55000, 'age': 28},
    {'name': 'Priya',  'dept': 'HR',      'salary': 42000, 'age': 25},
    {'name': 'Amit',   'dept': 'IT',      'salary': 61000, 'age': 31},
    {'name': 'Sunita', 'dept': 'Finance', 'salary': 48000, 'age': 29},
    {'name': 'Rohan',  'dept': 'HR',      'salary': 39000, 'age': 24},
    {'name': 'Kavya',  'dept': 'Finance', 'salary': 52000, 'age': 27},
]

dict_group = {}
for emp in employees:
    dept = emp['dept']
    if dept not in dict_group:
        dict_group[dept] = []
    dict_group[dept].append(emp)
    
for dept, emp_list in dict_group.items():
    print(dept, len(emp_list))

# hard for me, hint lene pde
```

```python
# C12:- Har department ki average salary calculate karo aur print karo. 
employees = [
    {'name': 'Rahul',  'dept': 'IT',      'salary': 55000, 'age': 28},
    {'name': 'Priya',  'dept': 'HR',      'salary': 42000, 'age': 25},
    {'name': 'Amit',   'dept': 'IT',      'salary': 61000, 'age': 31},
    {'name': 'Sunita', 'dept': 'Finance', 'salary': 48000, 'age': 29},
    {'name': 'Rohan',  'dept': 'HR',      'salary': 39000, 'age': 24},
    {'name': 'Kavya',  'dept': 'Finance', 'salary': 52000, 'age': 27},
]

dict_group = {}
for emp in employees:
    dept = emp['dept']
    if dept not in dict_group:
        dict_group[dept] = []
    dict_group[dept].append(emp)

for a, b in dict_group.items():
    avg = sum(e['salary'] for e in b)/len(b)   
    print(a, avg)

# hard for me, help leni pde
```
