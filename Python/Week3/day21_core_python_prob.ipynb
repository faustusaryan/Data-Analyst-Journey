## LEVEL 2 — Patterns

### L2-Q1 · Count karo by category  →  maps to P01
Ek orders list hai. Har `status` kitni baar aaya count karo.

```python
orders = [
    {"id": 1, "status": "delivered"},
    {"id": 2, "status": "pending"},
    {"id": 3, "status": "delivered"},
    {"id": 4, "status": "cancelled"},
    {"id": 5, "status": "pending"},
    {"id": 6, "status": "delivered"},
    {"id": 7, "status": "cancelled"},
    {"id": 8, "status": "pending"},
]

status_count = {}
for order in orders:
    status = order['status']
    if status not in status_count:
        status_count[status] = 1
    else:
        status_count[status] += 1
print(status_count)

#Alternative

from collections import Counter
status_count2 = Counter(order['status'] for order in orders)
print(status_count2)

#Alternative

status_count3 = {}
for order in orders:
    status = order['status']
    status_count3[status] = status_count3.get(status, 0) + 1
print(status_count3)
```

### L2-Q2 · Sum karo by category  →  maps to P02
Har `category` ka total `amount` sum karo.

```python
transactions = [
    {"category": "Food", "amount": 500},
    {"category": "Travel", "amount": 2000},
    {"category": "Food", "amount": 350},
    {"category": "Shopping", "amount": 1500},
    {"category": "Travel", "amount": 800},
    {"category": "Food", "amount": 200},
]

cat_total = {}
for t in transactions:
    cat = t['category']
    amount = t['amount']
    if cat not in cat_total:
        cat_total[cat] = amount
    else:
        cat_total[cat] += amount
print(cat_total)

#Alternative

cat_total2 = {}
for t in transactions:
    cat = t['category']
    amount = t['amount']
    cat_total2[cat] = cat_total2.get(cat, 0) + amount
print(cat_total2)
```

### L2-Q3 · Average by category  →  maps to P03
Ek step aur aage: pehle total + count, phir average nikalo.

**Important:** Direct average nahi nikal sakte ek loop mein — pehle total aur count alag track karo.

```python
scores = [
    {"subject": "Math", "score": 90},
    {"subject": "Science", "score": 75},
    {"subject": "Math", "score": 85},
    {"subject": "Science", "score": 80},
    {"subject": "Math", "score": 95},
    {"subject": "English", "score": 70},
    {"subject": "English", "score": 80},
]
# Step 1: total aur count alag-alag track karo
sub_total = {}
sub_count = {}
for s in scores:
    subject = s['subject']
    score = s['score']
    sub_count[subject] = sub_count.get(subject, 0) + 1
    sub_total[subject] = sub_total.get(subject, 0) + score

# Step 2: average nikalo
sub_avg = {}
for sub in sub_total:
    sub_avg[sub] = round(sub_total[sub]/sub_count[sub], 2)

print(sub_avg)
```
