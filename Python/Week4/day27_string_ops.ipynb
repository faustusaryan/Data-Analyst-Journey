```python
# Problem Statement
# Given a string, write a function most_frequent_char(text) that returns
# the character that appears the most number of times. Ignore spaces.
# If tie, any one of the top characters is acceptable.
# Input: 'data analyst'   →   Output: 'a'  (appears 4 times, spaces ignored)

def most_freq_char(text):
    freq = {}
    for a in text:
        if a == ' ':
            continue
        if a not in freq:
            freq[a] = 1
        else:
            freq[a] += 1
    return max(freq, key=lambda x: freq[x])
    
most_freq_char('data analyst')  
```

```python
# Q1. Given a list of messy name strings, clean each one:
#   • Remove leading/trailing whitespace
#   • Convert to title case (first letter of each word capitalized)
# Input:  [' amit sharma ', 'PRIYA SINGH', 'rahul verma', ' Sneha GUPTA ']
# Output: ['Amit Sharma', 'Priya Singh', 'Rahul Verma', 'Sneha Gupta']

name = [' amit sharma ', 'PRIYA SINGH', 'rahul verma', ' Sneha GUPTA ']
Names = []
for a in name:
    if a not in Names:
        Names.append(a.strip().title())
print(Names)

#Alternative

cleaned = [a.strip().title() for a in name]
print(cleaned)
```

```python
# Q2. Given a list of full names (first + last), extract them as list of tuples.
# Input:  ['Amit Sharma', 'Priya Singh', 'Rajesh Kumar', 'Ananya Roy']
# Output: [('Amit', 'Sharma'), ('Priya', 'Singh'), ('Rajesh', 'Kumar'), ('Ananya', 'Roy')]

names = ['Amit Sharma', 'Priya Singh', 'Rajesh Kumar', 'Ananya Roy']
split = [tuple(a.split()) for a in names]
print(split)
```

```python
# Q3. Given phone numbers in different formats, standardize to last 10 digits only.
# Input:  ['9876543210', '98-765-43210', '+91-9876543210', '91-9876543210']
# Output: ['9876543210', '9876543210', '9876543210', '9876543210']
# Approach: pehle saare non-digit characters hatao, phir last 10 digits lo.

contacts = ['9876543210', '98-765-43210', '+91-9876543210', '91-9876543210']
import re
phone = [re.sub(r'\D', '', a)[-10 : ] for a in contacts]
print(phone)
```

```python
# Q4. Convert salary strings to clean integers.
# Input:  ['₹75,000', '₹55,000', '₹60,000', '80000', '₹45,000']
# Output: [75000, 55000, 60000, 80000, 45000]
# Approach: ₹ sign aur commas hatao, phir int() mein convert karo.

amount = ['₹75,000', '₹55,000', '₹60,000', '80000', '₹45,000']
new_list = [int(a.replace('₹','').replace(',','').strip()) for a in amount]
print(new_list)

# Alternative

result = [int(re.sub(r'[^\d]', '', a)) for a in amount]
print(result)
```

```python
# Q5. Given: 
scores = [85, 92, 78, 60, 88, 75]

# a) Kya saare scores 50 se zyada hain?   → Expected: True
print(all(a > 50 for a in scores))

# b) Kya koi score 90 se zyada hai?       → Expected: True  (92 hai)
print(any(a > 90 for a in scores))

# c) Kya koi score 60 se kam hai?          → Expected: False (60 is NOT < 60)
print(any(a < 60 for a in scores))
```

```python
# Q6. Given: 
words = ['python', 'data', 'sql', 'pandas', 'analyst', 'excel']

# a) Sort by word length ascending (chhota pehle)
     # Expected: ['sql', 'data', 'excel', 'python', 'pandas', 'analyst']

print(sorted(words, key=len))

# b) Sort by word length descending (lamba pehle)
#      Expected: ['analyst', 'python', 'pandas', 'excel', 'data', 'sql']
# Note: python(6) aur pandas(6) same length — original order maintain hogi (stable sort).

print(sorted(words, key=len, reverse = True))
```

```python
# Q7. Given: 
mixed = [1, 'hello', 3.14, True, 42, 'world', 99, False, 7.5]
# Filter karo sirf integers — True aur False EXCLUDE karo.
# Output: [1, 42, 99]
# Important: bool Python mein int ka subclass hai.
#   isinstance(True, int) → True — isliye extra check zaroori hai.

filtered = [a for a in mixed if type(a) is int]
print(filtered)
```

```python
# Q9. Given emails, valid wale filter karo.
# Valid email ke 3 rules:
#   1. '@' hona chahiye
#   2. '@' se PEHLE kuch hona chahiye (at_pos > 0)
#   3. '@' ke BAAD '.' hona chahiye
# Input:  ['amit@gmail.com', 'priya@', '@domain.com', 'rahul.verma@company.co.in', 'invalid']
# Output: ['amit@gmail.com', 'rahul.verma@company.co.in']
# Note: '@domain.com' fail hoga — '@' pehle kuch nahi (at_pos = 0, not > 0).

emails = ['amit@gmail.com', 'priya@', '@domain.com', 'rahul.verma@company.co.in', 'invalid', 'xyz@co.']

cleaned = []
for a in emails:
    at_pos = a.find('@')
    dot_pos = a.find('.', at_pos)  # '@' ke baad pehla '.' dhundo
    if at_pos > 0 and dot_pos > at_pos and dot_pos < len(a) - 1:
        cleaned.append(a)

print(cleaned)
```

```python
# Q10. Given a list of employee dicts, print each in this format:
#   'Rajesh works in IT department with salary ₹75,000'
# Use f-strings with {:,} for comma-formatted salary.

# Expected Output:
# Rajesh works in IT department with salary ₹75,000
# Priya works in HR department with salary ₹55,000
# Amit works in Finance department with salary ₹60,000

employees = [
    {'name': 'Rajesh', 'dept': 'IT',      'salary': 75000},
    {'name': 'Priya',  'dept': 'HR',      'salary': 55000},
    {'name': 'Amit',   'dept': 'Finance', 'salary': 60000}
]

for a in employees:
    print(f"{a['name']} works in {a['dept']} department with salary ₹{a['salary']:,}")
```
