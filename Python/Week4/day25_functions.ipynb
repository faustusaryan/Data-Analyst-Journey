# Warmup:-

```python
# Task: {number: square} dict banao using dict comprehension
# (square matlab number ** 2)
numbers = [1, 2, 3, 4, 5]

sqr = {n: n**2 for n in numbers}
print(sqr)
```

```python
# Task: Sirf 60000 ya zyada (>= 60000) salary wale employees rakhein
salary_dict = {'Rajesh': 75000, 'Sunita': 55000,
               'Aakash': 60000, 'Meena': 80000, 'Deepak': 45000}

filtered = {name: salary for name, salary in salary_dict.items() if salary >= 60000}
print(filtered)
```

```python
# Task: {name: score} format ka dict banao using dict comprehension
students = [
    {'name': 'Amit',  'score': 85},
    {'name': 'Priya', 'score': 92},
    {'name': 'Rahul', 'score': 78}
]

result =  {a['name']: a['score'] for a in students}
print(result)
```

```python
# Task: Har status ki count wala dict banao
# (unique statuses ki list banao, phir har ek ki count karo)
statuses = ['Delivered', 'Cancelled', 'Delivered',
            'Pending', 'Delivered', 'Cancelled', 'Delivered']

counts = {word: statuses.count(word) for word in set(statuses)}
print(counts)
```

```python
# Task: Grade assign karo using dict comprehension
# Rule: 90+  → 'A'
#       75-89 → 'B'
#       60-74 → 'C'
#       below 60 → 'F'
marks = {'Amit': 85, 'Priya': 92, 'Rahul': 55,
         'Sneha': 73, 'Vikram': 41}

grades = {name: 'A' if mark >= 90 else 'B' if mark >= 75 else 'C' if mark >= 60 else 'F' for name, mark in marks.items()}
print(grades)
```

```python
# Exercise 1: Basic Function — Unique Elements (order maintain karo)
def get_unique(lst):
    seen = set()
    unique = []
    for a in lst:
        if a not in seen:
            seen.add(a)
            unique.append(a)
    return unique

# Test:
print(get_unique([1, 2, 2, 3, 1, 4, 3, 5]))
# Expected: [1, 2, 3, 4, 5]
```

```python
# Exercise 2: Function with Default Parameter — Sort by Key
def sort_by_key(data, key, reverse=False):
    return sorted(data, key = lambda x: x[key], reverse = reverse)

employees = [
    {'name': 'Rajesh', 'salary': 75000},
    {'name': 'Deepak', 'salary': 45000},
    {'name': 'Meena',  'salary': 80000}
]
print(sort_by_key(employees, 'name'))        # ascending
print(sort_by_key(employees, 'salary', True))  # descending
```

```python
# Exercise 3: Statistics Function — Mean, Median, Mode
from collections import Counter
def calculate_stats(numbers):
    mean = round(sum(numbers) / len(numbers), 2)
    
    numbers = sorted(numbers)   
    n = len(numbers)
    if n % 2 == 0:
        median = (numbers[n//2 - 1] + numbers[n//2]) / 2
    else:
        median = numbers[n//2]
    
    freq = Counter(numbers)
    max_count = max(freq.values())
    mode = [k for k, v in freq.items() if v == max_count]

    return {'mean': mean, 'median': median, 'mode': mode}

print(calculate_stats([75000, 55000, 60000, 80000, 45000, 60000, 50000]))
```
