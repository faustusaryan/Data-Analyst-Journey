```python
# depts = ['IT', 'HR', 'Finance', 'Marketing'] List comprehension se sirf 4 ya usse zyada letters wale dikhao
# (len >= 4) 

depts = ['IT', 'HR', 'Finance', 'Marketing']
new_lst = [lst for lst in depts if len(lst) >= 4]
print(new_lst)
```

```python
# names = ['Raj', 'Priya', 'Amit'] Loop se har naam UPPER mein print karo + uski length bhi 
# Format: 'RAJ - 3'

names = ['Raj', 'Priya', 'Amit']
for el in names:
    print(el.upper(), '-', len(el))
```

```python
# 1️⃣ Basic Function
# Function banao: welcome(name)
# Return kare: "Welcome <name>"

def welcome(name):
    return 'Welcome ' + name

welcome('Aryan')
```

```python
# 2️⃣ Parameters vs Arguments
# Function banao: multiply(a, b)
# Return kare: a * b

def multiply(a, b):
    return a * b
multiply(9, 8)
```

```python
# 3️⃣ Default Parameter
# Function banao: salary_after_tax(salary, tax=0.1)
# Return kare: salary after tax

def salary_after_tax(salary, tax = 0.1):
    return salary - salary * tax
salary_after_tax(45000, 0.05)
salary_after_tax(45000)
```

```python
# 4️⃣ If / Elif / Else inside function
# Function banao: grade(marks)

# Logic
# >= 80 → "A"
# >= 60 → "B"
# else → "C"

def grade(marks):
    if marks >= 80:
        return 'A'
    elif marks>= 60 and marks < 80:
        return 'B'
    else:
        return 'C'
        
grade(79)
```

```python
# 5️⃣ Function calling function
# Function 1: square(x) → x*x
# Function 2: cube(x) → x * square(x)

def square(x):
    return x*x

def cube(x):
    return x * square(x)

cube(3)
```

```python
# 6️⃣ *args (variable arguments)
# Function banao: total(*numbers)
# Return kare: sab numbers ka sum

def total(*nums):
    return sum(nums)

total(1, 2, 3, 4)
```
